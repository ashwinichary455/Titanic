VISUALIZATION

```python
#HISTROGRAM
for col in numerical_columns:
  plt.figure(figsize=(5,4))
  sns.histplot(df[col])
  plt.show()
```

```python
#BOXPLOT
for col in numerical_columns:
  plt.figure(figsize=(5,4))
  sns.boxplot(df[col])
  plt.show()
```

```python
#countplot
for col in numerical_columns:
  plt.figure(figsize=(5,4))
  sns.countplot(x=col,data=df)
  plt.show()
```

```python
#scatterplot
for col in numerical_columns:
  plt.figure(figsize=(5,4))
  sns.scatterplot(df[col])
  plt.show()
```

```python
#CORRELATION
plt.figure(figsize=(10,8))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```
